```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 as cv

img = cv.imread ('img/sst.jpg', 0)
rows, cols = img.shape

#PLot the original image 
plt.subplot (1, 2, 1) 
plt.title("Original") 
plt.imshow(img)

M = np.float32([[1, 0, 100], [0, 1, 50]])
transl = cv.warpAffine(img, M, (cols, rows))
                               
#PLot the translated image 
plt.subplot (1, 2, 2)
plt.title("Translated Image")
plt.imshow (transl)
                               
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2 as cv

img = cv.imread('img/sst.jpg', 0)
rows, cols = img.shape

#Plot the original image
plt.subplot(1, 2, 1) 
plt.title("original")
plt.imshow(img)

img_rotation = cv.warpAffine(img,
                              cv.getRotationMatrix2D((cols/2, rows/2),
                                                     30, 0.6), (cols, rows))

#Plot the rotated image
plt.subplot(1, 2, 2)
plt.title("Rotated Image")
plt.imshow(img_rotation)

cv.imwrite('rotation_out.jpg', img_rotation)
cv.waitKey (0)
cv.destroyAllWindows()
```

```python
# importing Libraries
import cv2 as cv

image = cv.imread('img/sst.jpg')

cv2.imshow('Original Image', image)
cv2.waitKey (0)

# Gaussian Blur
Gaussian = cv2.GaussianBlur(image,(7,7),0)
cv2.imshow('Gaussian Blurang', Gaussian)
cv2.waitKey (0)

#Median Blur
median = cv2.medianBlur(image, 5)
cv2. imshow('Median Blurring', median) 
cv2. waitKey (0)

# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2.imshow('Bilateral Blurring', bilateral)
cv2. waitKey(0)
cv2. destroyAllWindows()
```
